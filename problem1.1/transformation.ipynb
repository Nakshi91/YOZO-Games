```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
input_image = '1.jpg'
output_image = '2.jpg'
```

```python
img = cv2.imread(input_image)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_out = cv2.imread(output_image)
img_out = cv2.cvtColor(img_out, cv2.COLOR_BGR2RGB)
plt.imshow(img_out)
```

###### adjust the contrast for input image

```python
alpha = 1.1 # Contrast control (1.0-3.0)
beta = 0 # Brightness control (0-100)

adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
plt.imshow(adjusted)
```

###### do transformation for input image

```python

  
# Locate points of the documents or object which you want to transform
pts1 = np.float32([[0, 0], [900, 0], [0, 600], [900, 600]])
pts2 = np.float32([[0, 0], [990, 0], [0, 660], [990, 660]])
      
# Apply Perspective Transform Algorithm
matrix = cv2.getPerspectiveTransform(pts1, pts2)
result = cv2.warpPerspective(img, matrix, (990, 660))
# Wrap the transformed image
result
# plt.imshow('frame', frame) # Inital Capture
plt.imshow(result) # Transformed Capture
```
